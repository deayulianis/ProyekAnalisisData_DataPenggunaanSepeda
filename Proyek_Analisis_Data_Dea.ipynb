## Visualization & Explanatory Analysis

### Pertanyaan 1:
**Bagaimana pola penggunaan sepeda sepanjang hari?**

```python
# Mengelompokkan data berdasarkan jam
hourly_trend = all_df.groupby('hr').agg({
    'casual_hour': 'mean',
    'registered_hour': 'mean',
    'cnt_hour': 'mean'
}).reset_index()

# Membuat plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='hr', y='casual_hour', data=hourly_trend, label='Casual', marker='o')
sns.lineplot(x='hr', y='registered_hour', data=hourly_trend, label='Registered', marker='o')
sns.lineplot(x='hr', y='cnt_hour', data=hourly_trend, label='Total', marker='o')

# Menambahkan judul dan label
plt.title('Pola Penggunaan Sepeda Sepanjang Hari', fontsize=14)
plt.xlabel('Jam dalam Sehari', fontsize=12)
plt.ylabel('Jumlah Pengguna', fontsize=12)
plt.xticks(range(0, 24))  # Menampilkan semua jam (0-23)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

# Menampilkan plot
plt.show()
```

Kesimpulan:

Visualisasi menunjukkan bahwa penggunaan sepeda meningkat signifikan pada jam pagi (07:00 - 09:00) dan sore (17:00 - 19:00), yang kemungkinan besar berkaitan dengan jam berangkat dan pulang kerja/sekolah. Sementara itu, penggunaan sepeda paling rendah terjadi pada dini hari (00:00 - 05:00).

### Pertanyaan 2:
 **Bagaimana pengaruh musim terhadap jumlah peminjaman sepeda?**

```python
# Mengelompokkan data berdasarkan musim
season_trend = all_df.groupby('season_hour').agg({
    'casual_hour': 'mean',
    'registered_hour': 'mean',
    'cnt_hour': 'mean'
}).reset_index()

# Mengubah label musim
season_labels = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
season_trend['season_hour'] = season_trend['season_hour'].map(season_labels)

# Membuat plot
plt.figure(figsize=(10, 6))
sns.barplot(x='season_hour', y='cnt_hour', data=season_trend, palette='coolwarm')

# Menambahkan label dan judul
plt.title('Pengaruh Musim terhadap Jumlah Peminjaman Sepeda', fontsize=14)
plt.xlabel('Musim', fontsize=12)
plt.ylabel('Rata-rata Jumlah Peminjaman', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Menampilkan plot
plt.show()
```

Visualisasi terlihat jumlah peminjaman sepeda tertinggi terjadi pada musim gugur (Fall), diikuti oleh musim panas (Summer). Sementara itu, peminjaman sepeda paling rendah terjadi pada musim semi (Spring) dan musim dingin (Winter), kemungkinan karena kondisi cuaca yang kurang mendukung, seperti hujan atau suhu dingin yang ekstrem.
